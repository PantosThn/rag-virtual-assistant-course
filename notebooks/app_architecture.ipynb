```python
from graphviz import Digraph

def plot_langgraph_styled():
    dot = Digraph(format="png")
    dot.attr(rankdir="TB", size="6,8")  # Top-to-Bottom layout

    # Node styles
    dot.node("START", shape="circle", style="filled", color="#34c759")
    dot.node("_rewrite", label="‚úçÔ∏è Rewrite", shape="box")
    dot.node("_retrieve", label="üîé Retrieve", shape="box")
    dot.node("_grade", label="üß™ Grade", shape="box")
    dot.node("_web", label="üåê Web Search", shape="box")
    dot.node("_generate", label="üß† Generate", shape="box")
    dot.node("END", shape="circle", style="filled", color="#ff3b30", fontcolor="white")

    # Standard edges
    dot.edge("START", "_rewrite")
    dot.edge("_rewrite", "_retrieve")
    dot.edge("_retrieve", "_grade")

    # Conditional edges from _grade
    dot.edge("_grade", "_generate", label="have_ctx", style="dashed", color="#34c759")
    dot.edge("_grade", "_web", label="need_web", style="dashed", color="#007aff")

    # Rest of the flow
    dot.edge("_web", "_generate")
    dot.edge("_generate", "END")

    return dot

plot_langgraph_styled()
```
